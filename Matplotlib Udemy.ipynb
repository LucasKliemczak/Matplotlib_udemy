# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
theta = np.linspace(0,360,30)
y1 = np.sin(theta*np.pi/180)
y2 = np.cos(theta*np.pi/180)

x = np.linspace(0,10,11)
z1 = x**3
z2 = x**2

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(theta,y1)
plt.title('Z = cos'r'($\theta{}$)')
plt.tight_layout(pad=4.0)
plt.subplot(222)
plt.plot(theta,y2)
plt.title('Z = sin'r'($\theta{}$)')
plt.tight_layout(pad=4.0)

plt.subplot(223)
plt.plot(x,z1)
plt.title('Z = 'r'$x^{3}$')
plt.tight_layout(pad=4.0)
plt.subplot(224)
plt.plot(x,z2)
plt.title('Z = 'r'$x^{2}$')
plt.tight_layout(pad=4.0)
```

# GridSpec

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gsp
```

```python
x = np.linspace(0,5,250)
y1 = 2*x
y2 = (3*x**2) - np.exp(-(x-0.2)**2)
y3 = x**3 - x**2 - 1
```

```python
fig = plt.figure(figsize=(10,5), facecolor=(0.8,0.8,0.8))
gs = gsp.GridSpec(2,2)
ax1 = fig.add_subplot(gs[0,:])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])
ax1.plot(x,y1)
ax1.set_title('y1 = 2x')
ax2.plot(x,y2)
ax2.set_title('y2 = $3x^2 - e^{-(x-0.2)^2}$')
ax3.plot(x,y3)
ax3.set_title('y3 = $x^3 - x^2 - 1$')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gsp

x = np.linspace(0,5,250)
y1 = 2*x
y2 = (3*x**2) - np.exp(-(x-0.2)**2)
y3 = x**3 - x**2 - 1

fig = plt.figure(figsize=(10,5), dpi=120, facecolor = (0.8,0.8,0.8))
gs = gsp.GridSpec(2,2)

ax1 = fig.add_subplot(gs[0,:],facecolor = (0.9,0.9,0.9))
ax2 = fig.add_subplot(gs[1,0], facecolor = 'grey')
ax3 = fig.add_subplot(gs[1,1])

ax1.plot(x,y1, color=(1,0,1))
ax1.set_title('2x')

ax2.plot(x,y2)
ax2.set_title('y2 = $3x^2 - e^{-(x-0.2)^2}$')

ax3.plot(x,y2)
ax3.set_title('y3 = $x^3 - x^2 - 1$')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xp = np.linspace(-5,5,5)
yp = np.linspace(-5,5,5)

X, Y = np.meshgrid(xp, yp)

Z = np.sin(X) + np.cos(Y)

plt.contour(X,Y,Z,10, cmap='jet') # jet, rainbow, coolwarm
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xp = np.linspace(-5,5,5)
yp = np.linspace(-5,5,5)

X, Y = np.meshgrid(xp, yp)

Z = np.sin(X) * np.cos(Y)

plt.contourf(X,Y,Z,50, cmap='plasma') # jet, rainbow, coolwarm
plt.colorbar()
plt.show()
```

# Surface Plot

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X, Y = np.meshgrid(x, y)
Z = (X**2 + Y**2)
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
surfplot = ax.plot_surface(X,Y,Z, cmap = 'jet', alpha=0.9)
cset = ax.contourf(X,Y,Z, zdir='z', offset = 0, cmap = 'jet')

plt.colorbar(surfplot)
plt.show()
```
